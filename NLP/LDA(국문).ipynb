## 1. 데이터 업로드

```python
import pandas as pd
data = pd.read_csv("C:/Users/CIEL310/data/patent4.csv")
data.head()
```

```python
name = list(data['name'])
item = list(data['item'])
```

## 2. 데이터 전처리

```python
from konlpy.tag import Okt  
from konlpy.utils import pprint
okt=Okt()

name_noun = list(map(okt.nouns, name))
```

```python
print(name[1])
pprint(name_noun[1])
```

```python
item_noun = list()
for i in item :
    if type(i) != str :
        item_noun += ''
    else :
        item_noun.append(okt.nouns(i))
```

```python
print(item[1])
pprint(item_noun[1])
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stopwords = list()

with open('data/stopwords.txt', encoding='utf-8') as file :
    lines = file.readlines()
    for i in lines :
        stopwords.append(i.rstrip("\n"))
print(stopwords)
```

```python
def clear(word_tokenize) :
    result = list()
    for w in word_tokenize :
        if w not in stopwords :
            result.append(w)
    return result
```

```python
clear_name = list(map(clear, name_noun))
clear_item = list(map(clear, item_noun))
print(clear_name[1])
print(clear_item[1])
```

## 3.Word2Vec

```python
from gensim.models import Word2Vec

nmodel = Word2Vec(sentences=clear_name, vector_size=100, window=10, min_count=5, workers=10, sg=0)

model_result = nmodel.wv.most_similar("블록")
print(model_result)
```

```python
imodel = Word2Vec(sentences=clear_item, vector_size=100, window=10, min_count=5, workers=10, sg=0)

model_result = imodel.wv.most_similar("투표")
print(model_result)
```

## 4. LDA (전처리 적용)

```python
import gensim

dictionary = gensim.corpora.Dictionary(clear_item)
print('dictionary size : %d' % len(dictionary))
```

```python
dictionary[1]
```

```python
from gensim import corpora

corpus = [dictionary.doc2bow(text) for text in clear_item]
corpus[1] # 수행된 결과에서 두번째 문서 출력, 첫번째 문서의 인덱스는 0
print(len(corpus))
```

```python
NUM_TOPICS = 10
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)

topics = ldamodel.print_topics(num_words=4)
for topic in topics:
    print(topic)
```

```python
print(ldamodel.print_topics())
```

```python
import pyLDAvis.gensim

pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
pyLDAvis.display(vis)
```
