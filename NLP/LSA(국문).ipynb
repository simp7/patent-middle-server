## 데이터 전처리

```python
from konlpy.tag import Okt  
from konlpy.utils import pprint
okt=Okt()

name_noun = list(map(okt.nouns, name))
```

```python
print(name[1])
pprint(name_noun[1])
```

```python
item_noun = list()
for i in item :
    if type(i) != str :
        item_noun += ''
    else :
        item_noun.append(okt.nouns(i))
```

```python
print(item[1])
pprint(item_noun[1])
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stopwords = list()

with open('data/stopwords.txt', encoding='utf-8') as file :
    lines = file.readlines()
    for i in lines :
        stopwords.append(i.rstrip("\n"))
print(stopwords)
```

```python
def clear(word_tokenize) :
    result = list()
    for w in word_tokenize :
        if w not in stopwords :
            result.append(w)
    return result
```

```python
clear_name = list(map(clear, name_noun))
clear_item = list(map(clear, item_noun))
print(clear_name[1])
print(clear_item[1])
```

## LSA

```python
new_df = pd.DataFrame({'item':item}).fillna("")
new_df
```

```python
# 알파벳 이외 문자 제거
new_df['clean_doc'] = new_df['item'].str.replace("[^ㄱ-ㅎ|ㅏ-ㅣ|가-힣#]"," ")
```

```python
#길이 3 이하 문자 제거
new_df['clean_doc'] = new_df['clean_doc'].apply(lambda x: ' '.join([w for w in x.split()]))
new_df['clean_doc']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=1000,max_df=0.5,smooth_idf=True)

X = vectorizer.fit_transform(new_df['clean_doc'])
X.shape
```

```python
from sklearn.decomposition import TruncatedSVD #토픽모델링

svd_model = TruncatedSVD(n_components=5, algorithm = 'randomized', n_iter=100, random_state=12)
svd_model.fit(X)
print(svd_model.components_.shape)

#특허분석 보고서의 기술분류가 5가지이므로 5개로 가정
```

```python
svd_model.singular_values_
```

```python
terms = vectorizer.get_feature_names()
len(terms)
```

```python
n =20 #주요 단어 출력
components = svd_model.components_
for index, topic in enumerate(components):
    print('Topic %d: '%(index+1),[terms[i] for i in topic.argsort()[: -n -1:-1]])
```

```python
n =5 #주요 단어 출력 (중요도와 함께)
components = svd_model.components_
for index, topic in enumerate(components):
    print('Topic %d: '%(index+1),[(terms[i], topic[i].round(5)) for i in topic.argsort()[:-n - 1:-1]])
```
